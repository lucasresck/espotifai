# Create playlists dataset with Spotipy

```python
import pandas as pd
import spotipy
from tqdm.notebook import tqdm
from spotipy.oauth2 import SpotifyClientCredentials

auth_manager = SpotifyClientCredentials()
sp = spotipy.Spotify(auth_manager=auth_manager)
```

## Gather playlists

```python
playlists = []
q = sp.search('symphonic metal', limit=50, type='playlist')
playlists.extend(q['playlists']['items'])
while q['playlists']:
    q = sp.next(q['playlists'])
    playlists.extend(q['playlists']['items'])
    print(q['playlists']['next'])
```

```python
df = pd.DataFrame(playlists)
df.sample(5)
```

## For each playlist, gather track data

```python
# Only 50 for a while...
playlists = playlists[:50]

all_tracks = []
for i, playlist in tqdm(enumerate(playlists.copy()), total=len(playlists)):
    playlist = playlist.copy()
    print(playlist['name'])
    tracks = []
    q = sp.playlist_tracks(playlist['id'])
    tracks.extend(q['items'])
    while q['next']:
        q = sp.next(q)
        tracks.extend(q['items'])
    for key in list(playlist.keys()):
        playlist['playlist:'+key] = playlist.pop(key)
    for track in tracks:
        track.update(playlist)
    all_tracks.extend(tracks)
```

```python
tracks_df = pd.DataFrame(all_tracks)
tracks_df.sample(5)
```

## Convert track dict to DataFrame columns

```python
a = tracks_df.track.apply(pd.Series)
tracks_df[['track:'+column for column in a.columns]] = a
tracks_df.sample(5)
```

## Convert album dict to DataFrame columns

```python
a = tracks_df['track:album'].apply(pd.Series)
tracks_df[['album:'+str(column) for column in a.columns]] = a
tracks_df.sample(5)
```
